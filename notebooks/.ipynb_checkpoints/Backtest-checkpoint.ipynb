```python
import pandas as pd
from algotrading.backtest import Backtest
from sma_agent import SMA_Agent
import matplotlib.pyplot as plt
import numpy as np
import os
%matplotlib inline
```

```python
stocks = os.listdir("./Historical data/")
for i in np.linspace(0.01, 0.1, 10):
    for stock in stocks[1:7]:
        prices = pd.read_csv("./Historical data/"+stock)["Price"]
        agent = SMA_Agent(prices[0], 10, i, i)
        test = Backtest(agent)
        output = test.run(prices)
        fig, ax = plt.subplots()
        ax.plot(np.arange(len(prices)), output, color='green')
        ax.plot(np.arange(len(prices)), prices, color='red')

        ax.set(xlabel='Days', ylabel='INR',
               title=stock)
        ax.grid()
        plt.show()
```

```python
prices = pd.read_csv("Historical data/DABU Historical Data.csv")["Price"]
agent = SMA_Agent(10000, 10, 0.015, 0.015)
test = Backtest(agent)
output = test.run(prices)
fig, ax = plt.subplots()
ax.plot(np.arange(len(prices)), output, color='green')
ax.plot(np.arange(len(prices)), prices, color='red')

ax.set(xlabel='Days', ylabel='INR',
       title="DABUR Historical Data")
ax.grid()
plt.show()
```
