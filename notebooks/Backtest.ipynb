```python
import pandas as pd
import sys
sys.path.append('../')
from algotrading.backtest import Backtest
from algotrading.agents.sma_agent import SMA_Agent
import matplotlib.pyplot as plt
import numpy as np
import os
%matplotlib inline
```

```python
nifty_list = pd.read_csv("../Historical data/Nifty50list.csv")
```

```python
import quandl
quandl.ApiConfig.api_key = "FDEDsMbK1E2t_PMf7X3M"
for stock in nifty_list["Symbol"]:
    try:
        print(stock)
        prices = quandl.get('NSE/'+stock)["Close"]
        agent = SMA_Agent(5000, 10, 0.1, 0.1)
        test = Backtest(agent)
        output = test.run(prices)
        fig, ax = plt.subplots()
        ax.plot(np.arange(len(prices)), output, color='green')
        ax.plot(np.arange(len(prices)), prices, color='red')

        ax.set(xlabel='Days', ylabel='INR',
               title=stock)
        ax.grid()
        plt.show()
    except:
        print("An exception occurred while retrieving %s" % (stock))
```
