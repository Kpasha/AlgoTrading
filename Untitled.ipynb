```python
import pandas as pd
from backtest import Backtest
from sma_agent import SMA_Agent
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
prices = pd.read_csv("Historical data/DABU Historical Data.csv")["Price"]
```

```python
agent = SMA_Agent(10000, 10, 0.015, 0.015)
```

```python
test = Backtest(agent)
```

```python
output = test.run(prices)
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(len(prices)), output, color='orange')
ax.plot(np.arange(len(prices)), prices, color='orange')

ax.set(xlabel='Days', ylabel='INR',
       title='Monies')
ax.grid()
plt.show()
```
